3. Следующим этапом преобразуем признак **"Опыт работы"**. Его текущий формат - это: **<Опыт работы: n лет m месяцев, периоды работы в различных компаниях…>**. 

Из столбца нам необходимо выделить общий опыт работы соискателя в месяцах, новый признак назовем "Опыт работы (месяц)"

Для начала обсудим условия решения задачи:
* Во-первых, в данном признаке есть пропуски. Условимся, что если мы встречаем пропуск, оставляем его как есть (функция-преобразование возвращает NaN)
* Во-вторых, в данном признаке есть скрытые пропуски. Для некоторых соискателей в столбце стоит значения "Не указано". Их тоже обозначим как NaN (функция-преобразование возвращает NaN)
* В-третьих, нас не интересует информация, которая описывается после указания опыта работы (периоды работы в различных компаниях)
* В-четвертых, у нас есть проблема: опыт работы может быть представлен только в годах или только месяцах. Например, можно встретить следующие варианты:
    * Опыт работы 3 года 2 месяца…

В результате преобразования у вас должен получиться столбец, содержащий информацию о том, сколько месяцев проработал соискатель.
Выполните преобразование, ответьте на контрольные вопросы и удалите столбец **"Опыт работы"** из таблицы.


```python
def job_experience(arg=""):
    """Get job experience in months

    Args:
        arg (str or object, optional): job experience in years and/or months OR blank OR "Не указано". Defaults to "".

    Returns:
        _type_: int, nubmer of months OR None
    """
  
    if arg is np.nan or arg == "Не указано":
        return None
    else:
        job_exp_list = str(arg).lower().replace("опыт работы","").split()
        months_count = 0
        if len(job_exp_list) > 1:
            if 'месяц' in job_exp_list[1]:
                months_count = int(job_exp_list[0])
            elif job_exp_list[1] in ['год','года','лет']:
                months_count = int(job_exp_list[0]) * 12 
            if len(job_exp_list) > 3:           
                if 'месяц' in job_exp_list[3]:
                    months_count += int(job_exp_list[2])
        if months_count == 0:
            return None
    return months_count


hh_data['Опыт работы (месяц)'] = hh_data["Опыт работы"].apply(job_experience)

# Delete the column, unnecessary now
hh_data.drop('Опыт работы', axis=1, inplace=True)  
```

5. Рассмотрим поближе признаки **"Занятость"** и **"График"**. Сейчас признаки представляют собой набор категорий желаемой занятости (полная занятость, частичная занятость, проектная работа, волонтерство, стажировка) и желаемого графика работы (полный день, сменный график, гибкий график, удаленная работа, вахтовый метод).
На сайте hh.ru соискатель может указывать различные комбинации данных категорий, например:
* полная занятость, частичная занятость
* частичная занятость, проектная работа, волонтерство
* полный день, удаленная работа
* вахтовый метод, гибкий график, удаленная работа, полная занятость

Такой вариант признаков имеет множество различных комбинаций, а значит множество уникальных значений, что мешает анализу. Нужно это исправить!

Давайте создадим признаки-мигалки для каждой категории: если категория присутствует в списке желаемых соискателем, то в столбце на месте строки рассматриваемого соискателя ставится True, иначе - False.

Такой метод преобразования категориальных признаков называется One Hot Encoding и его схема представлена на рисунке ниже:
<img src=https://raw.githubusercontent.com/AndreyRysistov/DatasetsForPandas/main/ohe.jpg>
Выполните данное преобразование для признаков "Занятость" и "График", ответьте на контрольные вопросы, после чего удалите их из таблицы

```python
def get_category_cols(df, df_col):
    """Transform df_col to new set of columns with True/False values

    Args:
        df (_type_): dataframe
        df_col (_type_): column of dataframe

    Returns:
        _type_: edited dataframe with new columns
    """
    
    cat_set=set()
    
    # Get set of unique categories
    for row in df_col:
        cat_set.update(row.split(', '))
    
    # Create col for each category and fill it with True or False
    for cat in cat_set:
        df[cat] = df_col.apply(lambda x: cat in x)
    return df


hh_data = get_category_cols(hh_data,hh_data['Занятость'])
hh_data = get_category_cols(hh_data,hh_data['График'])

# Delete the columns, unnecessary now
hh_data.drop('Занятость', axis=1, inplace=True)  
hh_data.drop('График', axis=1, inplace=True)  
```

```python
# ** Ответы на вопросы
# Сколько людей ищут проектную работу или волонтёрство (в обоих столбцах стоит True)?
print(f"{(hh_data['проектная работа'] & hh_data['волонтерство']).value_counts()[1]} соискателей ищут проектную работу или волонтёрство")

# Сколько людей хотят работать вахтовым методом или с гибким графиком (в обоих столбцах стоит True)?
print(f"{(hh_data['вахтовый метод'] & hh_data['гибкий график']).value_counts()[1]} соискателей хотят работать вахтовым методом или с гибким графиком")
```

```python
currency_df = pd.read_csv('data/ExchangeRates.csv', sep=',')
currency_df.drop((set(currency_df.columns.values) - {'currency', 'date', 'close', 'proportion'}), axis=1, inplace=True) 

# Convert dates in both tables to the same form
currency_df['date'] = pd.to_datetime(currency_df['date']).dt.date
hh_data['Обновление резюме'] = pd.to_datetime(hh_data['Обновление резюме']).dt.date

# Extract the amount and currency of the desired salary in separate columns 
hh_data['ЗП цифра'] = hh_data['ЗП'].apply(lambda x: int(str(x).split()[0]))
hh_data['ЗП валюта'] = hh_data['ЗП'].apply(lambda x: str(x).split()[1])
hh_data.drop('ЗП', axis=1, inplace=True) # И удаляем исходный столбец

# Initialize currency conversion table
currency_dict = {'грн.': 'UAH',\
                 'USD': 'USD',\
                 'EUR': 'EUR', \
                 'бел.руб.': 'BYN', \
                 'KGS': 'KGS', \
                 'сум': 'UZS', \
                 'AZN': 'AZN', \
                 'KZT': 'KZT' \
                }

# Convert currency symbols in dataset
hh_data['ЗП валюта'] = hh_data['ЗП валюта'].apply(lambda x: x if x == 'руб.' else currency_dict[x])

# Merge tables and remove duplicate columns
hh_data = hh_data.merge(currency_df, how = 'left', left_on = ['ЗП валюта','Обновление резюме'], right_on = ['currency','date'])
hh_data.drop(['currency', 'date'], axis=1, inplace=True) 

# Fill NA values in the columns of the exchange rate and proportions for salaries in rubles. 
hh_data['close'] = hh_data['close'].fillna(1)
hh_data['proportion'] = hh_data['proportion'].fillna(1)

# Convert all salaries into rubles 
hh_data['ЗП (руб)'] = (hh_data['ЗП цифра'] * hh_data['close'] // hh_data['proportion']).apply(int) 
```

В какую сторону асимметрично логарифмическое распределение? Напишите об этом в комментарии к графику.
Найдите выбросы с помощью метода z-отклонения и удалите их из данных, используйте логарифмический масштаб. Давайте сделаем послабление на **1 сигму** (возьмите 4 сигмы) в **правую сторону**.

Выведите таблицу с полученными выбросами и оцените, с каким возрастом соискатели попадают под категорию выбросов?

```python
# Histogram on a logarithmic scale 
log_age = np.log(hh_data_dedupped['Возраст'] + 1)
mu = log_age.mean() # Среднее
sigma = log_age.std()
lower_bound = mu - 3*sigma # Left bound of 3 Sigma Method 
upper_bound = mu + 3*sigma # Right bound of 3 Sigma Method 

fig, ax = plt.subplots(figsize=(12, 6))
histplot = sns.histplot(log_age , bins=30)
histplot.set_title('Логарифмическое распределение возраста', fontsize=20, fontweight='bold', y=1.05)
histplot.figure.patch.set_facecolor('white')
histplot.axvline(mu, color='red', lw=2)
histplot.axvline(lower_bound, color='red', lw=2)
histplot.axvline(upper_bound, color='red', lw=2)
```

```python
upper_bound = mu + 4*sigma # Right bound  - relaxation by 1 sigma, i.e. 4 sigma to the right
outliers = hh_data_dedupped[(log_age < lower_bound) | (log_age > upper_bound)]
print(f'Число выбросов по методу z-отклонения: {outliers.shape[0]}')
print(outliers['Возраст'])

rows_to_delete = outliers.index
print(f'Число записей до удаления: {hh_data_dedupped.shape[0]}')
hh_data_dedupped.drop(index = rows_to_delete, inplace = True)
print(f'Результирующее число записей: {hh_data_dedupped.shape[0]}')
```
